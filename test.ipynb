```python
import cv2 
import os 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def data_bal(path):
    data = os.listdir(path)
    df = []
    for file_name in data: 
        if '.jpg' in file_name:
            img_p = os.path.join(path, file_name)
            df.append(img_p)
            img = plt.imread(img_p)
    return df

df = data_bal('./Tesis')
```

```python
# img = plt.imread(df[0])
# plt.imshow(img)

def preproccess_image(df):
    df_1 = []
    for file_path in df:
        img = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)
        img = cv2.bitwise_not(img)
        img[img > 180] = 255 # Asignar valor 255 a píxeles más oscuro
        img[img < 160] = 0   # Asignar valor 0 a píxeles más sclaros
        df_1.append(img)
    return df_1

```

```python
# Llamar a la función preproccess_image con la lista de rutas de imágenes df
df_processed = preproccess_image(df)

# Mostrar la primera imagen preprocesada
img_processed = df_processed[2]
plt.imshow(img_processed, cmap='gray')  # Mostrar la imagen en escala de grises
plt.title("Imagen Preprocesada")
plt.axis(False)
plt.show()
```
